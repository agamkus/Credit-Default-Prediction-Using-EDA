### 4. Data Quality Check and Missing Value:

##### Percentage of missing values in all the columns:

```python
round(100 * app_data.isnull().sum()/app_data.shape[0], 2).sort_values(ascending=False)
```

##### Removing columns with more than 50% missing values.

```python
app_data = app_data.loc[:, round(100 * app_data.isnull().sum()/app_data.shape[0], 2) <50]
```

##### There are few more columns which are having around 47% missing values.Since these are near to 50% ,we are removing these columns as well:

```python
app_data = app_data.loc[:, round(100 * app_data.isnull().sum()/app_data.shape[0], 2) <47]
```

##### Now, shape of the dataset changed:

```python
app_data.shape
```

##### Filtering columns with low missing percentage values; let's say 13% and creating a new DataFrame "less_miss_df":

```python
less_miss_df = app_data.loc[:, (round(100 * app_data.isnull().sum()/app_data.shape[0], 2) <14) & round(100 * app_data.isnull().sum()/app_data.shape[0], 2)>0]
```

##### Getting data types of colums:

```python
less_miss_df.dtypes
```

##### As we can see there is only one column "NAME_TYPE_SUITE" with data type 'object', we can replace missing values with mode of the column:

```python
app_data['NAME_TYPE_SUITE'] = app_data['NAME_TYPE_SUITE'].fillna(app_data['NAME_TYPE_SUITE'].mode()[0])
```

##### Now, it is showing 0 null values in the column "NAME_TYPE_SUITE":

```python
app_data['NAME_TYPE_SUITE'].isnull().sum()
```

##### Now, treating missing values for continuos columns which are filterred with less percentage of missing values:

##### Selecting all the columns which are with continuous data type:

```python
cols = list(less_miss_df.select_dtypes(include=np.number).columns)
```

```python
less_miss_df[cols].describe(percentiles=[0,0.25,0.50,0.75,0.95,0.97,0.99])
```

##### As per the abobe description, few columns are with outliers. So, we are selecting 3 of them and replacing null values with Median:

```python
app_data['OBS_30_CNT_SOCIAL_CIRCLE'] = app_data['OBS_30_CNT_SOCIAL_CIRCLE'].fillna(app_data['OBS_30_CNT_SOCIAL_CIRCLE'].median())
app_data['OBS_60_CNT_SOCIAL_CIRCLE'] = app_data['OBS_60_CNT_SOCIAL_CIRCLE'].fillna(app_data['OBS_60_CNT_SOCIAL_CIRCLE'].median())
app_data['AMT_REQ_CREDIT_BUREAU_QRT'] = app_data['AMT_REQ_CREDIT_BUREAU_QRT'].fillna(app_data['AMT_REQ_CREDIT_BUREAU_QRT'].median())
```

##### Now, selecting 2 columns which are not with outliers and replacing null values with mean of the column:

```python
app_data['AMT_GOODS_PRICE'] = app_data['AMT_GOODS_PRICE'].fillna(app_data['AMT_GOODS_PRICE'].mean())
app_data['EXT_SOURCE_2'] = app_data['EXT_SOURCE_2'].fillna(app_data['EXT_SOURCE_2'].mean())
```

##### Getting data type of all the columns:

```python
app_data.dtypes
```

##### As per the above mentioned data types, the columns "DAYS_REGISTRATION" is with float value, but it should be integer value, so, changing its data type to int:

```python
app_data['DAYS_REGISTRATION'] = app_data['DAYS_REGISTRATION'].astype('int64')
```

### Correlation:

##### To find correlation, we are selecting few columns which are impacting the analysis. Below are the columns:

##### Matrix Plot for Target 0:

```python
corr_target_0 = target_0[columns_for_corr]
corr_mat_0 = corr_target_0.corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr_mat_0, annot=True)
plt.title("Correlation Matrix for Target 0 Columns")
plt.show()
```

##### Top 10 correlations for Target 0:

```python
corr_mat_0 = corr_target_0.corr()
corr = corr_mat_0.where(np.triu(np.ones(corr_mat_0.shape), k=1).astype(np.bool))
corr = corr.abs().unstack().reset_index()
corr.columns = ['VAR1', 'VAR2', 'CORR']
corr.dropna(subset=['CORR'], inplace=True)
corr.sort_values(by='CORR', ascending=False).head(10)
```
